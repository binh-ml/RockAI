```python
import sys
sys.path.append('/home/thanhbinh/code')
from butils.base import print_progress
from butils.imutils import load_image_file, write_img_cv2, resize

import os
from glob import glob
from PIL import Image
```

```python
def resize_images(input_dir, output_dir, size):
    os.makedirs(output_dir, exist_ok=True)
    filepaths = [f for f in glob(os.path.join(input_dir, '**'), recursive=True)
                 if os.path.isfile(f) and os.path.splitext(f)[-1].lower() in ['.jpg', '.png']]
    num_files = len(filepaths)
    for i, fin in enumerate(filepaths):        
        print_progress(f'[{i+1}/{num_files}] {fin}')
        fout = os.path.join(output_dir, os.path.relpath(fin, input_dir))        
#         img = Image.open(fin)
#         img.resize(size).save(fout)   
        img = load_image_file(fin)
        img = resize(img, size, keep_aspect=False, crop=False)
        write_img_cv2(img, fout)
```

```python
size = (224, 224)
```

```python
resize_images('data/original/No_RA/', f'data/resize_{size[0]}_{size[1]}/No_RA/', size)
```

```python
resize_images('data/original/RA/', f'data/resize_{size[0]}_{size[1]}/RA/', size)  
```
