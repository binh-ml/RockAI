```python
import os
from PIL import Image
```

```python
DATA_DIR = 'data/original/'
```

```python
RA_filenames = [f for f in os.listdir(os.path.join(DATA_DIR, 'RA'))
                if os.path.isfile(os.path.join(DATA_DIR, 'RA', f)) and 
                   os.path.splitext(f)[-1].lower() in ['.jpg', '.jpeg', '.png']]
print(f'Number of Rock-Art images: {len(RA_filenames)}')

No_RA_filenames = [f for f in os.listdir(os.path.join(DATA_DIR, 'No_RA'))
                   if os.path.isfile(os.path.join(DATA_DIR, 'No_RA', f)) and 
                      os.path.splitext(f)[-1].lower() in ['.jpg', '.jpeg', '.png']]
print(f'Number of Non-Rock-Art images: {len(No_RA_filenames)}')
```

```python
RA_image_size = []
for f in RA_filenames:
    img = Image.open(os.path.join(DATA_DIR, 'RA', f))
    RA_image_size.append(img.size)
print(set(RA_image_size))
```

```python
for s in set(RA_image_size):
    print(s[0]*s[1])
```

```python
No_RA_image_size = []
for f in No_RA_filenames:
    img = Image.open(os.path.join(DATA_DIR, 'No_RA', f))
    No_RA_image_size.append(img.size)
print(set(No_RA_image_size))
```

```python
for s in set(No_RA_image_size):
    print(s[0]*s[1])
```
